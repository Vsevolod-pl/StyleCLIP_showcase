```python
import torch
import numpy as np
from matplotlib import pyplot as plt

import clip
from models.stylegan2.model import Generator
#import e4e
```

```python
def draw_img(img):
    with torch.no_grad():
        img = img[0].cpu()
        scale = img.max()-img.min()
        d = img.min()/scale
        return plt.imshow(img.transpose(0,2).transpose(1,0)/scale-d)
```

```python
stylegan_size = 1024
stylegan_path = 'stylegan2-ffhq-config-f.pt'

generator = Generator(stylegan_size, 512, 8).cuda()
generator.load_state_dict(torch.load(stylegan_path)["g_ema"], strict=False)
latent_avg = generator.mean_latent(4096)

model, preprocess = clip.load("ViT-B/32", device="cuda")
```

```python
gd = torch.load('./global_dirs_relevance_vector5.pt')
```

```python
cos_sim2 = torch.nn.CosineSimilarity(dim=2)
```

```python
neutral = 'face with normal hair'
target =  'face with grey hair'

with torch.no_grad():
    text_neut = model.encode_text(clip.tokenize(neutral).cuda()).cpu()
    text_tar = model.encode_text(clip.tokenize(target).cuda()).cpu()
    dt = (text_tar-text_neut)[0].to(dtype=torch.float32).cuda()
```

```python
R = cos_sim2(gd, dt) # torch.tensordot(gd, dt.reshape(1,1,-1))#
#R /= R.abs().max()
```

```python
plt.hist(R.flatten().cpu().numpy(), bins=100);
```

```python
lat = torch.randn(1, 512).cuda()
with torch.no_grad():
    img, lat, _ = generator([lat], input_is_latent=False, randomize_noise=False, return_latents=True, truncation=0.7, truncation_latent=latent_avg,)

draw_img(img)
```

```python
plt.figure(figsize=(15, 9))
i = 0
mx_it = torch.abs(R).max().item()
for beta in torch.linspace(0.05, mx_it*0.8, 3):
    mask = torch.zeros_like(R)+1
    mask[torch.abs(R) < beta] = 0
    gamma = torch.norm(mask*R)
    for alpha in torch.linspace(-10, 10, 5):
        i += 1
        lat2 = lat+(mask*R*alpha/gamma).cuda()

        with torch.no_grad():
            img2, _ = generator([lat2], input_is_latent=True, randomize_noise=False)
        plt.subplot(3, 5, i)
        plt.axis('off')
        draw_img(img2)
```
